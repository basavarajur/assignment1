
Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is
a vowel, False otherwise.

```python
def is_vowel(ch):
    vowels = ('a', 'e', 'i', 'o', 'u')
    if ch not in vowels:
        return False
    return True



print (is_vowel('a'))
print (is_vowel('e'))
print (is_vowel('i'))
print (is_vowel('o'))
print (is_vowel('u'))
print (is_vowel('b'))
print (is_vowel('c'))
print (is_vowel(1))
```
