
Write a Python program to implement your own myfilter() function which works exactly like
Python's built-in function filter()

```python

def myfilter(anyfunc, sequence):
    result = []
    for item in sequence:
        if anyfunc(item):
            result.append(item)
    return result


def ispositive(x):
    if (x <= 0): 
        return False 
    else:
        return True

list= range(-10,11)
print ("Filter only positive Integers in list are: "  + str(myfilter(ispositive, list)))

```
