
Implement a function longestWord() that takes a list of words and returns the longest one.

```python
def find_longest_word(words_list):
    word_len = []
    for n in words_list:
        word_len.append((len(n), n))
    word_len.sort()
    return word_len[-1][1]

longest_word = find_longest_word(["kokila", "Exercises", "basavaraju", "Backend"])
print(longest_word)
```
